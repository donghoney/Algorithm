```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def obj_fun(a, b, x, y, m):              #objective function 함수
  return sum((a + b * x[i] - y[i]) ** 2 for i in range(m)) / (2 * m)

path = "data.csv"
data = np.genfromtxt(path, delimiter=',')

x_data = data[:, 0]
y_data = data[:, 1]                                        

TH0 = np.arange(-30, 30, 0.1)          #θ0 range 설정
TH1 = np.arange(-30, 30, 0.1)          #θ1 range 설정
TH0, TH1 = np.meshgrid(TH0, TH1)       #surface 그리기 위해 base grid 생성

JTH = obj_fun(TH0, TH1, x_data, y_data, len(x_data))    #objective function 사용해 값 받아오기

fig = plt.figure()
ax = fig.gca(projection='3d')    #3d axes 생성
ax.plot_surface(TH0, TH1, JTH, rstride=1, cstride=1, alpha=0.2)  #surface 그리기
ax.set_xlabel('θ0')
ax.set_ylabel('θ1')
ax.set_zlabel('J') 

plt.show()
```
